```python
# Exercício 1
# Criar uma função que converta de Celsius para Farenheit

def celsiusToFarenheit (celsius):
    return (celsius * 1.8) + 32

def farenheitToCelsius (farenheit):
    return (farenheit - 32) / 1.8
```

```python
# Exercício 2
# Implementar uma função que retorne verdadeiro se o número for primo (falso caso contrário). Testar de 1 a 100

def isPrime (number):
    if number == 0:
        return False

    if number == 1:
        return True

    numbers = []
    for i in range(1, number + 1):
        if number % i == 0:
            numbers.append(i)
    
    return len(numbers) == 2

for i in range(1, 100):
    print('Numero analisado {}'.format(i))
    print('É primo {}'.format(isPrime(i)))
```

```python
# Exercício 3
# Implementar uma função que receba uma lista de listas de comprimentos quaisquer e retorne uma lista de uma dimensão

def concatLists (list1, list2):
    return list1 + list2

concatLists([1, 2, 3], [4, 5, 6])
```

```python
# Exercício 5
# Escreva uma função que:

# Receba uma frase como parâmetro.
# Retorne uma nova frase com cada palavra com as letras invertidas.

def revertPhrase (phrase):
    return phrase[::-1]

revertPhrase('Eu sou terrivel')
```
